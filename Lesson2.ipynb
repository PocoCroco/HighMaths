```python
x = []
y1 = []
y2 = []
for i in range(-500, 501):
    x_ = i/100
    x.append(x_)
    y1.append(np.sqrt(25-x_**2))
    y2.append(-np.sqrt(25-x_**2))
plt.plot(x, y1, color = 'b')
plt.plot(x, y2, color = 'b')
plt.xlabel('x')
plt.ylabel('y')
plt.axis('scaled')
```

### Задание 4

Решить систему неравенств:

exp(x) + x∙(1 – y) = 1  
y = x2 – 1

```python
from scipy.optimize import fsolve

def func(p):
    x, y = p
    return (x**2 - 1 - y, np.exp(x) + x * (1 - y) - 1)

x1, y1 =  fsolve(func, (-2, 1))
x2, y2 =  fsolve(func, (1, 4))
              
print (x1, y1)
print (x2, y2)
```
